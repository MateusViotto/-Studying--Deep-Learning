# Introdução

### O que é Deep Learning? E Machine Learning?

Deep Learning é um ramo de aprendizado de máquina baseado em um conjunto de algoritmos que tentam modelar abstrações de alto nível de dados usando um grafo profundo com várias camadas de processamento, compostas de várias transformações lineares e não lineares. 

Machine lerning is turning things (data) into numbers and <b>finding patterns</b> in those numbers. (The computer does this part)

### Machine Learning vs Deep Leaning

Machine learning é melhor com dados estruturados, deep learning é melhor com dados não estruturados.

**Traditional programming:** INPUT -> RULES -> OUTPUT (Makes the output based in the rules)    
**Machine learning algorithm:** INPUT -> OUTPUT -> RULES (Figures out the rules based in the output)

# Why use Machine Learning or Deep Learning?

*Why not?*

Para problemas complexos não conseguimos pensar em todas as regras, logo utilizamos ML. 

*If you can build a simple rule-based systemm that doesnn't require machine learning, do that.*

### What deep learning is good for? ✅

- **Problems with long list of rules:** when the traditional approach fails, machine learning or deep learning may help  
- **Continually changing environments:** deep learning can adapt ('learn') to new scenarios.  
- **Discovering insights within large collections of data:** can you imagine trying to hand-craft rules for what 101 different kinds of food look like?  

### What deep learning is not good for? 🚫

- **When you need explainability:** the patterns learned by a deep lerning model are typically uninterpretable by a human  
- **When the traditional approach is a better option:** if you can accomplish what you need with a simple rule-based system   
- **When errors are unacceptable:** since the outputs of deep learning model aren't always predictable  
- **When you don't have much data:** deep learning models usually require a fairly large amount of data to produce great results  

# What are neural networks?

Em ciência da computação e campos relacionados, redes neurais artificiais ou redes neuronais artificiais são modelos computacionais inspirados pelo sistema nervoso central de um animal que são capazes de realizar o aprendizado de máquina bem como o reconhecimento de padrões.

INPUT -> NUMERAL ENCODING -> LEARNS REPRESENTATION (patterns/features/weights) -> REPRESENTATION OUTPUTS -> OUTPUTS

- **Input Layer:** data goes in here  
- **Hidden layer(s):** learns patterns in data  
- **Output layer:** outputs learned representation or prediction probabilities
